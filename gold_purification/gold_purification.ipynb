```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_absolute_error
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats as st
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import RandomizedSearchCV #подбор гипер параметров
from sklearn.metrics import make_scorer
pd.set_option('display.max_columns', None)
```

```python
def dummy_reg(train_for_ml, final_df, rougher_df, target_1, target_2):
    
    dummy_regr = DummyRegressor(strategy="mean")
    
    y1 = target_1
    y2 = target_2
    x = train_for_ml
    
    rougher_dummy_regr = dummy_regr.fit(x, y1)
    final_dummy_regr = dummy_regr.fit(x, y2)
   
    roug = smape(rougher_df['rougher.output.recovery'], rougher_dummy_regr.predict(rougher_df.drop('rougher.output.recovery', axis=1))) * 0.25
    final = smape(final_df['final.output.recovery'], final_dummy_regr.predict(final_df.drop('final.output.recovery', axis=1))) * 0.75
    result_smape = roug + final
    

    print(result_smape)
```

```python
def smape(value_true, value_predict):
    return 100/len(value_true) * np.sum(2 * np.abs(value_predict - value_true) / (np.abs(value_true) + np.abs(value_predict)))
```

```python
def recovery(c, f, t):
    # C — доля золота в концентрате после флотации/очистки;
    # F — доля золота в сырье/концентрате до флотации/очистки;
    # T — доля золота в отвальных хвостах после флотации/очистки.
    return ((c * (f - t)) / (f * (c - t))) * 100
```

```python
def craft_pipeline(train_for_ml, target_1, target_2, smape=smape, rfr_flag=True):

    
    
    if rfr_flag:
        model = RandomForestRegressor(random_state=12345)
        param_grid = {
        'randomforestregressor__n_estimators': range(30, 110, 10),
        'randomforestregressor__min_samples_split': (2, 3, 4),
        'randomforestregressor__min_samples_leaf': (1, 2, 3, 4)
        }
        pipline_model = make_pipeline(model)
    else:
        scaler = StandardScaler()
        scaler.fit(train_for_ml) 
        train_for_ml = scaler.transform(train_for_ml)
        
        model = SGDRegressor(random_state=12345)
        
        param_grid = {
            'sgdregressor__learning_rate': ['adaptive'],
            'sgdregressor__alpha' : 10.0**-np.arange(1,7),
            'sgdregressor__l1_ratio':[.05, .15, .5, .7, .9, .95, .99, 1],
            'sgdregressor__max_iter': [3000]
        }
        
            
           
       # 'sgd__subsample'    : [0.5, 0.2, 0.1],
       # 'sgd__n_estimators' : [100, 1000, 1500],
       # 'sgd__max_depth'    : range(15,20)

        #print(model.get_params().keys())
        pipline_model = make_pipeline(model)

    gs = RandomizedSearchCV(
        pipline_model, 
        param_distributions=param_grid, 
        scoring=make_scorer(smape, greater_is_better=False), 
        n_jobs=-1,
        cv=3,
        random_state=12345
    )

    gs_model1 = gs.fit(train_for_ml, target_1)
    
    gs_model1_best_score = gs_model1.best_score_ * -1
    gs_model1_best_params = gs_model1.best_params_
    print(f'best_score: {gs_model1_best_score}')
    print(f'best_params: {gs_model1_best_params}')
    
    gs_model2 = gs.fit(train_for_ml, target_2)
    
    gs_model2_best_score = gs_model2.best_score_ * -1
    gs_model2_best_params = gs_model2.best_params_
    print(f'best_score: {gs_model2_best_score}')
    print(f'best_params: {gs_model2_best_params}')   
    
    
    return gs_model2.best_estimator_, gs_model2.best_estimator_
```

```python
def good_col(rougher_df, train_for_ml, col_string='rougher.output.recovery'):
    rougher_target = rougher_df[col_string]
    rougher_df = rougher_df.drop(col_string, axis=1)
    rougher_df = rougher_df[train_for_ml.columns]
    rougher_df[col_string] = rougher_target
    return rougher_df
```
